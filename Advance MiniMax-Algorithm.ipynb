```python
import random
```

```python
#Function to check if the player wins?                               
def win(board,player):
  #Defining the character that needs to be compared
  if player==0:
    c='O'
  else:
    c='X'

  #Checking the match
  # Checking for Rows
  for row in range(3) :
    if (board[row][0] == board[row][1] and board[row][1] == board[row][2]) :
      if board[row][0]==c:
        return 1

  # Checking for Columns for X or O victory.
  for col in range(3) :
    if (board[0][col] == board[1][col] and board[1][col] == board[2][col]) :
      if board[0][col]==c:
        return 1

  # Checking for Diagonals for X or O victory.
  if (board[0][0] == board[1][1] and board[1][1] == board[2][2]) :
    if board[0][0]==c:
      return 1

  if (board[0][2] == board[1][1] and board[1][1] == board[2][0]) :
    if board[0][2]==c:
      return 1

  return 0
```

```python
#game stopping Function (checking if there are no blank spaces)
def stop(board):
  again=0
  for row in range(3):
    for col in range(3):
      if board[row][col]=='-':
        again=1
  if again==0:
    return True
  else :
    return False
```

```python
#handles the turn of user (Makes necessary changes on the board)
def userturn(board):
  while True:
    x=int(input("Select X coordinate : "))
    y=int(input("Select Y coordinate : "))
    if board[x][y]=='-':
      board[x][y]='O'
      break
    else:
      print('Wrong Position selected')
```

```python
#Actual Minimax Algorithm function
def minimax(board,player):                
  #The base conditions
  if win(board,1):     #Machine wins
    return 1
  if win(board,0):     #User wins
    return -1
  if stop(board):      #All spaces are filled(No one wins)
    return 0

  #If the board has empty spaces
  #If the Maximizer(Machine) is playing
  if player==1:
    best=-100
    for i in range(3):
      for j in range(3):
        if (board[i][j]=='-'):
          board[i][j] = 'X'
          best = max(best, minimax(board,0))
          board[i][j] = '-'
    return best
  #If the Minimizer(Assumed Player) is playing
  if player==0:
    best=100
    for i in range(3):
      for j in range(3):
        if (board[i][j]=='-'):
          board[i][j] = 'O'
          best = min(best, minimax(board,1))
          board[i][j] = '-'
    return best
```

```python
#Function that selects the best position from the available positions based on their minimax values
def best(entries):
  best=[]         #for storing the coordinates of Machine win
  medium=[]       #for storing the coordinates of neutral
  worst=[]        #for storing the coordinates of user win
  for row in range(len(entries)):
    if(entries[row][2])==1:
      best.append([entries[row][0],entries[row][1]])

  for row in range(len(entries)):
    if(entries[row][2])==0:
      medium.append([entries[row][0],entries[row][1]])

  for row in range(len(entries)):
    if(entries[row][2])==-1:
      worst.append([entries[row][0],entries[row][1]])

  #randomly shuffeling the entries of coordinates to insert randomness
  random.shuffle(best)
  random.shuffle(medium)
  random.shuffle(worst)

  #Generating the final answer list
  ans_list=best           #by default ans_list is the set of best outcome
  if len(best)==0:        #if no cordintes lead to best outcome go for neutral
    ans_list=medium
  if len(medium)==0:      #if no coordinates lead to neutral accept the failure
    ans_list=worst

  #Dealing with situations when all the coordinates are either of neutral or failure
  immediate=imediate_loose(board)        #getting the coordinates in which will lead to the user win in next immediate step
  # print('immediate:',immediate,end='')
  # print('ans_list:',ans_list,end='')

  if immediate in ans_list:              #if immediate coordinates are present in the ans_list then use them to slow down the opponent success
    return immediate[0],immediate[1]
  else:
    return ans_list[0][0],ans_list[0][1] #else return the default coordinates
```

```python
#function for getting the coordinates in which will lead to the user win in next immediate step
def imediate_loose(board):
  for i in range(3):
    for j in range(3):
      if (board[i][j]=='-'):
          board[i][j] = 'O'
          if win(board,0):
            board[i][j] = '-'
            return [i,j]
          board[i][j] = '-'
```

```python
def print_board(board):
  for row in board:
    print(row)
```

```python
board=[['-','-','-'],
       ['-','-','-'],
       ['-','-','-']]

print_board(board)

#Main Function
while True:
  #User's Turn
  print("Your Turn")
  userturn(board)
  if win(board,0):
    print("You Won !")
    break
  print_board(board)

  if(stop(board)):
    print("Game Over")
    break

  #Machine's Turn
  entries=[]
  for row in range(3):
    for col in range(3):
      entry=[]
      if (board[row][col]=='-'):
        board[row][col] = 'X'
        value=minimax(board,0)
        board[row][col] = '-'
        entry.append(row)
        entry.append(col)
        entry.append(value)
        entries.append(entry)

  #print(entries)
  x,y=best(entries)
  print("\n")
  print("Machine-coordinates",x,y,end='\t')
  board[x][y]='X'
  print('\n')
  print('Machine Made its move')
  print_board(board)
  print('\n')


  if win(board,1):
    print("Machine Won !")
    break

  if(stop(board)):
    print("Game Over")
    break





```
